```python
# 字幕・フィラー削除設定

print("=== 字幕設定 ===\n")


max_chars_input = input(
    "1行あたりの最大文字数の目安を入力してください(1〜60、Enterでデフォルト値20): "
)

max_chars_per_line = (
    int(max_chars_input)
    if max_chars_input.strip()
    else 20
)

max_chars_per_line = min(
    max(max_chars_per_line, 1),
    60
)


# 1行×2行を基本としたブロック上限
limit = max_chars_per_line * 2


block_max_input = input(
    f"\nブロック分割の目安文字数を入力してください(1〜120、Enterでデフォルト値{limit}): "
)

block_max_chars = (
    int(block_max_input)
    if block_max_input.strip()
    else limit
)

block_max_chars = min(
    max(block_max_chars, 1),
    120
)


if block_max_chars > limit:
    print(
        f"入力値が上限({limit})を超えていたため、{limit}に自動調整しました。"
    )
    block_max_chars = limit


# ----------------------------------
# 字幕表示行数
# ----------------------------------

print("\n字幕表示を選んでください")
print("1: 1行表示")
print("2: 2行表示（デフォルト）")

line_mode_input = input(
    "番号を入力してください(Enterでデフォルト値2): "
)

subtitle_lines = (
    "1"
    if line_mode_input.strip() == "1"
    else "2"
)


# ----------------------------------
# Gemini校正
# ----------------------------------

print("\nGeminiによる校正を行いますか？")
print("1: する（デフォルト）")
print("2: しない")

gemini_input = input(
    "番号を入力してください(Enterでデフォルト値1): "
)

use_gemini = (
    "2"
    if gemini_input.strip() == "2"
    else "1"
)


print("\nフィラー削除の強さを選んでください")
print("1: 弱  2: 中(デフォルト)  3: 強")

filler_input = input(
    "番号を入力してください(Enterでデフォルト値2): "
)

filler_map = {
    "1": "弱",
    "2": "中",
    "3": "強"
}

filler_level = filler_map.get(
    filler_input.strip(),
    "中"
)


print("\n--- 設定内容 ---")
print(f"1行あたり最大文字数: {max_chars_per_line}")
print(f"ブロック分割の目安文字数: {block_max_chars}")
print(f"字幕表示: {subtitle_lines}行")
print(f"Gemini校正: {'する' if use_gemini == '1' else 'しない'}")
print(f"フィラー削除レベル: {filler_level}")
```

## ③ 共通関数

```python
# 共通関数

import re


def remove_fillers(text, level="中"):
    """
    フィラーを除去する

    ルール:
    ・えーっと、えっと、えーと、えー、ええとは全レベルで削除
    ・あ、え、は文頭のみ削除
    ・ああ、ええ、は誤削除防止のため残す
    ・そうですねは強レベルのみ削除
    """

    filler_words = {
        "弱": [
            "えーっと",
            "えーと",
            "ええと",
            "えっと",
            "えー"
        ],

        "中": [
            "えーっと",
            "えーと",
            "ええと",
            "えっと",
            "えー",
            "あの",
            "その",
            "まぁ",
            "まあ",
            "うーん"
        ],

        "強": [
            "えーっと",
            "えーと",
            "ええと",
            "えっと",
            "えー",
            "あの",
            "その",
            "まぁ",
            "まあ",
            "うーん",
            "そうですね",
            "なんか",
            "こう"
        ]
    }

    # 長いフィラーから先に削除
    for filler in sorted(
        filler_words.get(level, filler_words["中"]),
        key=len,
        reverse=True
    ):
        text = text.replace(filler, "")

    # 「あ、」「え、」は文頭のみ削除
    # 「ああ、」「ええ、」には反応しない
    text = re.sub(r"^あ[、。]\s*", "", text)
    text = re.sub(r"^え[、。]\s*", "", text)

    # 余分な空白整理
    text = re.sub(r"\s+", " ", text).strip()

    return text


def sec_to_srt_time(sec):
    """
    秒(float) → SRTタイムコード
    """

    hrs = int(sec // 3600)
    mins = int((sec % 3600) // 60)
    secs = int(sec % 60)
    msecs = int(round((sec - int(sec)) * 1000))

    if msecs >= 1000:
        msecs -= 1000
        secs += 1

    return f"{hrs:02d}:{mins:02d}:{secs:02d},{msecs:03d}"


def chunks_to_srt_text(chunks):
    """
    processed_segments → SRT文字列
    """

    lines = []

    for i, chunk in enumerate(chunks, start=1):
        lines.append(
            f"{i}\n"
            f"{sec_to_srt_time(chunk['start'])} --> {sec_to_srt_time(chunk['end'])}\n"
            f"{chunk['text']}"
        )

    return "\n\n".join(lines)

# ==========================================
# Gemini校正用 共通関数
# ==========================================


def extract_text_for_gemini(chunks):
    """
    Geminiへ送信する本文のみを作成
    タイムコード情報は送信しない
    """

    lines = []

    for i, c in enumerate(chunks, start=1):

        lines.append(
            f"{i}\n{c['text']}"
        )

    return "\n\n".join(lines)



def parse_gemini_result(text):
    """
    Gemini結果から番号ごとの本文を取得
    """

    blocks = re.split(
        r"\n\s*\n",
        text.strip()
    )

    result = {}

    for block in blocks:

        lines = block.strip().split("\n")

        if len(lines) >= 2:

            try:

                num = int(
                    lines[0].strip()
                )

                body = "".join(
                    lines[1:]
                )

                result[num] = body


            except:

                pass


    return result

# 字幕1行あたり最大文字数
target_max_chars = max_chars_per_line
```

## ④ Whisper文字起こし

```python
# ==========================================
# ④ Whisper文字起こし（句読点保持用）
# ==========================================

print("=== Whisper文字起こし開始 ===")


import whisper
import torch
import gc
import copy


# ==========================================
# 設定
# ==========================================

device = "cuda" if torch.cuda.is_available() else "cpu"

print("使用デバイス:", device)


# ==========================================
# Whisperモデル読み込み
# ==========================================

model = whisper.load_model(
    "large-v3",
    device=device
)


# ==========================================
# 音声文字起こし
# ==========================================

result = model.transcribe(
    filename,
    language="ja",
    fp16=(device == "cuda"),
    condition_on_previous_text=False
)


# ==========================================
# 句読点保持用保存
# Whisper直後の結果を安全に保持
# ==========================================

whisper_result_raw = copy.deepcopy(result)


print(
    "Whisper文字起こし完了"
)


print(
    "segment数:",
    len(whisper_result_raw["segments"])
)


# ==========================================
# 句読点確認
# ==========================================

print(
    "句読点確認:"
)


count = 0


for s in whisper_result_raw["segments"]:

    if "。" in s["text"] or "、" in s["text"]:

        print(
            s["text"]
        )

        count += 1


print(
    "句読点segment数:",
    count
)


# ==========================================
# Whisperモデル解放
# ==========================================

del model

gc.collect()

if torch.cuda.is_available():
    torch.cuda.empty_cache()


print(
    "=== Whisperモデル解放完了 ==="
)
```

## ⑤ WhisperXアライメント

```python
# ==========================================
# ⑤ WhisperXアライメント
# ==========================================

print("=== WhisperXアライメント開始 ===")


import whisperx
import torch
import gc


# ==========================================
# 設定
# ==========================================

device = "cuda" if torch.cuda.is_available() else "cpu"

print("使用デバイス:", device)


# ==========================================
# 音声読み込み(WhisperX用)
# ==========================================

audio = whisperx.load_audio(filename)


# ==========================================
# WhisperXアライメントモデル読み込み
# ==========================================

model_a, metadata = whisperx.load_align_model(
    language_code="ja",
    device=device
)


# ==========================================
# アライメント実行
# Whisper文字起こし結果に時間情報を付与
# 本文はWhisper、時間情報のみWhisperXを使用
# ==========================================

whisper_result_aligned = whisperx.align(
    whisper_result_raw["segments"],
    model_a,
    metadata,
    audio,
    device,
    return_char_alignments=True
)


# ==========================================
# 確認
# ==========================================

print(
    "=== WhisperXアライメント完了 ==="
)


word_count = sum(
    len(s.get("words", []))
    for s in whisper_result_aligned["segments"]
)


print(
    "WhisperX word数:",
    word_count
)


# ==========================================
# WhisperXモデル解放
# ==========================================

del model_a

gc.collect()

if torch.cuda.is_available():
    torch.cuda.empty_cache()


print(
    "=== WhisperXモデル解放完了 ==="
)
```

## ⑥ WhisperX SRT生成

```python
# ==========================================
# ⑥ WhisperX SRT生成
# Whisper本文 + WhisperX文字タイムコード版
#
# 役割：
# ・WhisperX chars生成
# ・句点「。」単位のみ分割
# ・文字タイムコード保持
#
# ※字幕の改行・文字数調整は後工程で実施
# ==========================================

print("=== WhisperX SRT生成開始 ===")


if "whisper_result_raw" not in locals():

    raise NameError(
        "whisper_result_raw がありません。"
        "⑤ Whisper文字起こしを先に実行してください。"
    )


if "whisper_result_aligned" not in locals():

    raise NameError(
        "whisper_result_aligned がありません。"
        "⑤.1 WhisperXアライメントを先に実行してください。"
    )


# ==========================================
# 秒 → SRT時間
# ==========================================

def sec_to_srt_time(sec):

    h = int(sec // 3600)

    m = int((sec % 3600) // 60)

    s = int(sec % 60)

    ms = int(
        (sec - int(sec)) * 1000
    )


    return (
        f"{h:02d}:"
        f"{m:02d}:"
        f"{s:02d},"
        f"{ms:03d}"
    )






# ==========================================
# 句点「。」のみで分割
# ==========================================

def split_text_by_period(
    char_timing
):

    """
    句点単位で字幕ブロック生成

    ・「。」は話者境界候補として保持
    ・読点「、」では分割しない
    ・文字数制限なし
    ・chars同期維持
    """


    blocks = []

    current_chars = []



    for char_info in char_timing:


        current_chars.append(
            char_info
        )


        if char_info["char"] == "。":


            blocks.append(
                {
                    "text":
                        "".join(
                            c["char"]
                            for c in current_chars
                        ),

                    "start":
                        current_chars[0]["start"],

                    "end":
                        current_chars[-1]["end"],

                    "chars":
                        current_chars
                }
            )


            current_chars = []



    # ======================================
    # 最後に句点がない残り
    # ======================================

    if current_chars:


        blocks.append(
            {
                "text":
                    "".join(
                        c["char"]
                        for c in current_chars
                    ),

                "start":
                    current_chars[0]["start"],

                "end":
                    current_chars[-1]["end"],

                "chars":
                    current_chars
            }
        )


    return blocks

# ==========================================
# Whisper本文 + WhisperX時間から字幕生成
# ==========================================

blocks = []


raw_segments = (
    whisper_result_raw["segments"]
)


aligned_segments = (
    whisper_result_aligned["segments"]
)



for idx, raw_segment in enumerate(
    raw_segments
):


    if idx >= len(aligned_segments):

        continue



    aligned_words = (
        aligned_segments[idx]
        .get("chars", [])
    )



    if not aligned_words:

        continue



    # ==================================
    # WhisperX文字タイムコード作成
    # ==================================

    char_timing = aligned_words


    if not char_timing:

        continue



    # ==================================
    # 句点単位で分割
    # ==================================

    segment_blocks = split_text_by_period(
        char_timing
    )



    blocks.extend(
        segment_blocks
    )



print(
    "字幕ブロック数:",
    len(blocks)
)



# ==========================================
# ⑦へ渡すデータ作成
# ==========================================

processed_segments = []


for i, block in enumerate(
    blocks,
    start=1
):


    processed_segments.append(
        {
            "id":
                i,

            "start":
                block["start"],

            "end":
                block["end"],

            "text":
                block["text"],

            "chars":
                block.get(
                    "chars",
                    []
                )
        }
    )



# ==========================================
# SRTプレビュー生成
# ==========================================

srt_blocks = []



for i, block in enumerate(
    processed_segments,
    start=1
):


    srt_blocks.append(
        f"{i}\n"
        f"{sec_to_srt_time(block['start'])} --> "
        f"{sec_to_srt_time(block['end'])}\n"
        f"{block['text']}"
    )



raw_srt = "\n\n".join(
    srt_blocks
)



print(
    "\n=== WhisperX SRT生成完了 ==="
)


print(
    "字幕数:",
    len(processed_segments)
)


print(
    "processed_segments作成完了"
)


print(
    "\n=== SRTプレビュー ===\n"
)


print(raw_srt)
```

## ⑦ カスタムフィラー

**標準のフィラー削除機能で削除されない言葉を、追加で指定できます。**

以下のコード内の「CUSTOM_FILLERS = """」より下に、削除したい言葉を1行に1つずつ入力してください。

入力した言葉は、字幕生成時に自動で削除されます。

## ⑨ フィラー除去（Python）
文字起こし結果から、不要なフィラーを自動で削除します。

```python
if "processed_segments" not in locals():
    raise NameError("processed_segments が見つかりません。")


print("フィラー除去中...")


# ==================================
# chars基準フィラー除去
# startのみ更新
# endは維持
# ==================================

def remove_fillers_with_chars(segment):

    chars = segment.get(
        "chars",
        []
    )


    # charsがない場合は従来処理

    if not chars:

        text = segment["text"]

        segment["text"] = remove_fillers(
            text,
            filler_level
        )

        return segment



    # ==================================
    # 元charsから文字列作成
    # ==================================

    original_text = "".join(
        c["char"]
        for c in chars
    )



    # ==================================
    # フィラー除去後文字列取得
    # ==================================

    filtered_text = remove_fillers(
        original_text,
        filler_level
    )



    # ==================================
    # 削除なしの場合
    # ==================================

    if filtered_text == original_text:

        segment["text"] = original_text

        segment["chars"] = chars

        return segment



    # ==================================
    # 削除された文字位置を特定
    # ==================================

    new_chars = []

    old_index = 0


    for new_char in filtered_text:


        while old_index < len(chars):

            if chars[old_index]["char"] == new_char:

                new_chars.append(
                    chars[old_index]
                )

                old_index += 1

                break


            old_index += 1



    # ==================================
    # new_charsから最終テキスト生成
    # ==================================

    filtered_text = "".join(
        c["char"]
        for c in new_chars
    )



    # ==================================
    # text / chars更新
    # ==================================

    segment["text"] = filtered_text

    segment["chars"] = new_chars



    # ==================================
    # startのみ更新
    # endは変更禁止
    # ==================================

    if new_chars:

        segment["start"] = (
            new_chars[0]["start"]
        )



    return segment



# ==================================
# 実行
# ==================================

for segment in processed_segments:

    remove_fillers_with_chars(
        segment
    )



current_srt = chunks_to_srt_text(
    processed_segments
)



print("\n=== フィラー除去完了 ===")

print(current_srt)


print(
    "\n次の処理用に processed_segments を維持しました。"
)
```

## ⑩ 字幕の改行・分割調整

```python

# ==========================================
# 指定文字数・行数への最終整形
#
# WhisperX文字タイムコード利用版
#
# 目的:
# ・⑥から来た長い字幕ブロックを整形
# ・自然な位置で分割
# ・charsタイムコード維持
# ・①のユーザー設定を利用
# ・単語途中分割禁止
# ・句点・読点優先分割
# ・Janome境界利用
# ==========================================


import re
from janome.tokenizer import Tokenizer


tokenizer = Tokenizer()



# ==========================================
# 設定取得
# ①から受け取り
# ==========================================

if "max_chars_per_line" not in locals():

    raise NameError(
        "max_chars_per_line がありません。①字幕設定を先に実行してください。"
    )


if "block_max_chars" not in locals():

    raise NameError(
        "block_max_chars がありません。①字幕設定を先に実行してください。"
    )



target_max_chars = max_chars_per_line

block_limit = block_max_chars



print(
    "=== 最終字幕整形設定 ==="
)

print(
    f"1行最大文字数: {target_max_chars}"
)

print(
    f"ブロック最大文字数: {block_limit}"
)


# ==========================================
# 分割スコア設定
# ==========================================

DISTANCE_WEIGHT = 1.0


# ==========================================
# 助詞候補
# ==========================================

TARGET_PARTICLES = {
    "ので",
    "から",
    "けど",
    "けれど",
    "けれども",
    "ですが",
    "けども"
}



# ==========================================
# charsから分割時間取得
# ==========================================

def get_split_time_by_chars(
    chars,
    split_pos,
    start_sec,
    end_sec,
    full_text=None
):

    if not chars:

        total = split_pos if split_pos else 1

        return (
            start_sec
            +
            (
                end_sec
                -
                start_sec
            )
            *
            0.5
        )


    if split_pos < len(chars):

        return chars[split_pos]["start"]


    return chars[-1]["end"]

# ==========================================
# Janome境界取得
# 品詞詳細保持版
# ==========================================

def find_word_boundaries(text):


    result = []

    pos = 0



    for token in tokenizer.tokenize(text):


        surface = token.surface


        start = pos


        pos += len(surface)



        pos_detail = (
            token.part_of_speech
            .split(",")
        )


        part = pos_detail[0]


        sub_part = (
            pos_detail[1]
            if len(pos_detail) > 1
            else ""
        )


        conjugation = (
            pos_detail[5]
            if len(pos_detail) > 5
            else ""
        )



        result.append(
            {
                "start": start,

                "pos": pos,

                "surface": surface,

                # 名詞・動詞など
                "part": part,

                # 固有名詞・一般・非自立など
                "sub_part": sub_part,

                # 活用情報
                "conjugation": conjugation,

                # Janome完全情報
                "pos_detail": pos_detail
            }
        )



    return result


# ==========================================
# 分割候補取得
# 改良版
#
# ・優先順位制御
# ・副詞境界低下
# ・敬称境界対策
# ・複合名詞境界制御
# ・引用符境界追加
# ==========================================

def get_split_candidates(text):


    candidates = []

    center = len(text) / 2



    # ----------------------------------
    # 括弧範囲取得
    #
    # 内部の分割候補を除外するため
    # ----------------------------------

    protected_ranges = []


    brackets = [
        ("『", "』"),
        ("（", "）"),
        ("(", ")")
    ]


    for start_mark, end_mark in brackets:


        start = 0


        while True:

            start = text.find(
                start_mark,
                start
            )


            if start == -1:

                break


            end = text.find(
                end_mark,
                start + 1
            )


            if end == -1:

                break


            protected_ranges.append(
                (
                    start,
                    end + 1
                )
            )


            start = end + 1



    # ----------------------------------
    # 候補追加用
    #
    # 括弧内部は禁止
    # ----------------------------------

    def add_candidate(
        pos,
        priority,
        type_name
    ):


        # ----------------------------------
        # 括弧内部は禁止
        # ----------------------------------

        for start, end in protected_ranges:

            if start < pos < end:

                return



        # ----------------------------------
        # 禁止表現チェック
        #
        # 境界前後を確認
        #
        # 例:
        # してく｜ださい
        #
        # → してください
        #
        # 分割禁止
        # ----------------------------------

        forbidden_patterns = [
            "ください",
            "下さい",
            "いただく",
            "いただき",
            "いただいて",
            "いただける"
        ]


        left_text = text[:pos]

        right_text = text[pos:]


        for word in forbidden_patterns:

            word_len = len(word)


            for i in range(
                max(0, pos - word_len + 1),
                pos + 1
            ):


                if text[i:i + word_len] == word:


                    if (
                        i < pos
                        and
                        pos < i + word_len
                    ):

                        return



        candidates.append(
            {
                "pos": pos,

                "priority": priority,

                "type": type_name,

                "distance":
                    abs(
                        pos - center
                    )
            }
        )


    # ------------------------------
    # ① 句点
    # priority 300
    # ------------------------------

    for m in re.finditer(
        "。",
        text
    ):


        add_candidate(
            m.end(),
            300,
            "句点"
        )



    # ------------------------------
    # ② 読点
    # priority 250
    # ------------------------------

    for m in re.finditer(
        "、",
        text
    ):


        add_candidate(
            m.end(),
            250,
            "読点"
        )



    # ------------------------------
    # ③ 指定助詞
    #
    # 後 priority 150
    # 前 priority 100
    # ------------------------------

    pos = 0


    tokens = list(
        tokenizer.tokenize(text)
    )


    for token in tokens:


        surface = token.surface


        start_pos = pos


        pos += len(surface)



        if surface in TARGET_PARTICLES:


            add_candidate(
                pos,
                150,
                "指定助詞後"
            )


            add_candidate(
                start_pos,
                100,
                "指定助詞前"
            )
    # ------------------------------
    # ④ 引用符境界
    #
    # 「温泉女子旅」
    # 名前と「温泉」
    #
    # 前後を優先候補にする
    # priority 220
    # ------------------------------

    for m in re.finditer(
        "「",
        text
    ):


        add_candidate(
            m.start(),
            220,
            "引用符前"
        )



    for m in re.finditer(
        "」",
        text
    ):


        add_candidate(
            m.end(),
            220,
            "引用符後"
        )



    # ------------------------------
    # ⑤ Janome境界
    # ------------------------------

    boundaries = find_word_boundaries(text)



    for i, b in enumerate(boundaries):


        pos = b["pos"]

        start = b["start"]

        surface = b["surface"]

        part = b["part"]

        sub_part = b.get(
            "sub_part",
            ""
        )



        # ----------------------------------
        # 次の単語情報
        # ----------------------------------

        next_part = ""

        next_surface = ""


        if i + 1 < len(boundaries):


            next_token = boundaries[i + 1]


            next_part = next_token["part"]

            next_surface = next_token["surface"]



        # ----------------------------------
        # 敬称境界チェック
        # ----------------------------------

        honorific_words = [
            "さん",
            "様",
            "さま",
            "君",
            "くん"
        ]


        skip_honorific = False



        if next_surface in honorific_words:

            skip_honorific = True



        for honorific in honorific_words:

            if surface.endswith(honorific):

                skip_honorific = True



        # ----------------------------------
        # 副詞後
        # ----------------------------------

        if part == "副詞":


            add_candidate(
                pos,
                10,
                "副詞後"
            )



        # ----------------------------------
        # 名詞の前
        #
        # 複合名詞内部を除外
        #
        # 例:
        # 観光｜資源 → 禁止
        # 動画｜編集 → 禁止
        #
        # ただし、
        # 助詞を挟む場合は許可
        #
        # 例:
        # 地域の｜観光
        # 温泉が｜自慢
        # ----------------------------------

        if part == "名詞":


            # ----------------------------------
            # 複合名詞内部チェック
            #
            # 直前も名詞で、かつ連続している場合は
            # 複合名詞内部と判断する
            #
            # その場合、
            # 名詞前 priority 200 の候補を作らない
            # ----------------------------------

            is_compound_continuation = False


            # ----------------------------------
            # 「の」直後チェック
            #
            # AのB のような所有格は強く結合するため
            # 「の」の直後で分割しない
            #
            # 例:
            # 地域の｜観光資源 → 禁止
            # 旅館の｜魅力 → 禁止
            # ----------------------------------

            is_after_no = False



            if i > 0:


                prev = boundaries[i - 1]


                if (
                    prev["part"] == "助詞"
                    and
                    prev["surface"] == "の"
                    and
                    prev["pos"] == start
                ):

                    is_after_no = True



            if i > 0:


                prev = boundaries[i - 1]


                if (
                    prev["part"] == "名詞"
                    and
                    prev["pos"] == start
                ):

                    is_compound_continuation = True



            if (
                not skip_honorific
                and
                not is_compound_continuation
                and
                not is_after_no
            ):


                if len(surface) == 1:


                    add_candidate(
                        start,
                        130,
                        "1文字名詞前"
                    )


                else:


                    add_candidate(
                        start,
                        200,
                        "名詞前"
                    )


        # ----------------------------------
        # 動詞の前
        # ----------------------------------

        if part == "動詞":


            if sub_part != "非自立":


                add_candidate(
                    start,
                    180,
                    "動詞前"
                )


            else:


                add_candidate(
                    start,
                    60,
                    "補助動詞前"
                )



        # ----------------------------------
        # 助詞・助動詞境界
        # ----------------------------------

        if part in [
            "助詞",
            "助動詞"
        ]:


            add_candidate(
                pos,
                80,
                "助詞助動詞境界"
            )



        # ----------------------------------
        # 名詞・動詞境界後
        #
        # 名詞→名詞は禁止
        # ----------------------------------

        if part in [
            "名詞",
            "動詞"
        ]:


            if next_part in [
                "名詞",
                "動詞"
            ]:


                # 複合名詞内部は禁止
                if not (
                    part == "名詞"
                    and
                    next_part == "名詞"
                ):


                    add_candidate(
                        pos,
                        20,
                        "名詞動詞境界後"
                    )



    # ----------------------------------
    # Janome一般境界
    # ----------------------------------

    for b in boundaries:


        add_candidate(
            b["pos"],
            50,
            "Janome一般境界"
        )



    return candidates


# ==========================================
# 最適分割位置選択
# priority直接評価版
# ==========================================

def select_split_position(
    text,
    candidates
):


    if not candidates:

        return None



    scored = []



    for c in candidates:


        pos = c["pos"]



        # 極端な分割は禁止

        if (
            pos < 5
            or
            pos > len(text) - 5
        ):

            continue



        score = (

            c["priority"]

            -

            c["distance"] * DISTANCE_WEIGHT

        )



        scored.append(
            (
                score,
                pos
            )
        )



    if not scored:

        return None



    scored.sort(
        reverse=True
    )


    return scored[0][1]


# ==========================================
# 行折返し
# priority直接評価版
# ==========================================

def wrap_text(
    text,
    max_width
):


    if len(text) <= max_width:

        return [
            text
        ]



    candidates = get_split_candidates(
        text
    )



    # 2文字程度の超過を許容

    hard_limit = (
        max_width + 2
    )



    min_second_line = 4



    min_pos = max(
        1,
        len(text) - hard_limit
    )


    max_pos = min(
        hard_limit,
        len(text) - min_second_line
    )



    scored = []



    for c in candidates:


        pos = c["pos"]



        # 文字数範囲外は除外

        if (
            pos < min_pos
            or
            pos > max_pos
        ):

            continue



        distance = abs(
            pos - max_width
        )



        score = (

            c["priority"]

            -

            distance

        )



        scored.append(
            (
                score,
                pos
            )
        )



    if scored:


        scored.sort(
            reverse=True
        )


        cut = scored[0][1]



    else:


        cut = min(
            max_width,
            len(text) - 1
        )



    line1 = (
        text[:cut]
        .strip()
    )


    line2 = (
        text[cut:]
        .strip()
    )



    return [
        line1,
        line2
    ]

# ==========================================
# 再帰分割
# ==========================================

def split_block(
    text,
    start_sec,
    end_sec,
    chars=None
):


    # ==================================
    # 句点による強制分割
    #
    # 「。」は改行ではなく字幕ブロック分割
    # 話者転換対策
    # ==================================

    period_positions = [
        m.end()
        for m in re.finditer(
            "。",
            text
        )
    ]


    if period_positions:


        split_pos = period_positions[0]


        # 極端な分割は禁止

        if (
            split_pos > 1
            and
            split_pos < len(text) - 1
        ):


            split_sec = get_split_time_by_chars(
                chars,
                split_pos,
                start_sec,
                end_sec,
                text
            )


            if (
                split_sec - start_sec >= 0.8
                and
                end_sec - split_sec >= 0.8
            ):


                if chars:


                    chars1 = chars[:split_pos]

                    chars2 = chars[split_pos:]


                else:


                    chars1 = None

                    chars2 = None



                result1 = split_block(
                    text[:split_pos],
                    start_sec,
                    split_sec,
                    chars1
                )


                result2 = split_block(
                    text[split_pos:],
                    split_sec,
                    end_sec,
                    chars2
                )


                return (
                    result1
                    +
                    result2
                )



    # 先に行折返し結果を取得
    # lines未定義問題修正

    lines = wrap_text(
        text,
        target_max_chars
    )



    soft_limit = int(
        block_limit * 0.9
    )



    need_split = (

        len(text)
        >
        soft_limit

    )



    # 文字数・行数が問題なければ終了

    if (
        not need_split
        and
        len(lines)
        <= 2
    ):

        return [
            (
                start_sec,
                end_sec,
                lines,
                chars
            )
        ]



    candidates = get_split_candidates(
        text
    )



    split_pos = select_split_position(
        text,
        candidates
    )


    if split_pos is None:


        return [
            (
                start_sec,
                end_sec,
                lines,
                chars
            )
        ]




    part1 = text[:split_pos]


    part2 = text[split_pos:]


    if (
        not part1
        or
        not part2
    ):

        return [
            (
                start_sec,
                end_sec,
                lines,
                chars
            )
        ]

    split_sec = get_split_time_by_chars(
        chars,
        split_pos,
        start_sec,
        end_sec,
        text
    )



    # 短すぎる字幕は禁止

    if (
        split_sec - start_sec
        <
        0.8
        or
        end_sec - split_sec
        <
        0.8
    ):

        return [
            (
                start_sec,
                end_sec,
                lines,
                chars
            )
        ]



    if chars:


        chars1 = chars[:split_pos]

        chars2 = chars[split_pos:]


    else:


        chars1 = None

        chars2 = None



    result1 = split_block(
        part1,
        start_sec,
        split_sec,
        chars1
    )


    result2 = split_block(
        part2,
        split_sec,
        end_sec,
        chars2
    )


    return (
        result1
        +
        result2
    )



# ==========================================
# processed_segments適用
# ==========================================

def format_segments(
    segments
):


    output = []

    new_id = 1



    for seg in segments:


        text = (
            seg["text"]
            .strip()
        )


        chars = seg.get(
            "chars",
            []
        )



        result = split_block(
            text,
            seg["start"],
            seg["end"],
            chars
        )



        for (
            start,
            end,
            lines,
            c
        ) in result:


            output.append(
                {
                    "id": new_id,

                    "start": start,

                    "end": end,

                    "text":
                        "\n".join(
                            lines
                        ),

                    "chars": c
                }
            )


            new_id += 1



    return output



# ==========================================
# 実行
# ==========================================

if "processed_segments" not in locals():

    raise NameError(
        "processed_segments がありません。"
    )



formatted_segments = format_segments(
    processed_segments
)



current_srt = chunks_to_srt_text(
    formatted_segments
)



print(
    "=== 最終整形完了 ==="
)


print(
    f"総ブロック数: {len(processed_segments)}"
)


print(
    f"1行最大文字数: {target_max_chars}"
)


print(
    f"ブロック最大文字数: {block_limit}"
)


print(
    "\n--- 整形後プレビュー ---"
)


print(
    current_srt
)

```

## ⑪1行字幕へ分割

```python
# ==========================================
#  1行字幕へ分割
#
# subtitle_line_mode == "1" の時のみ実行
# ==========================================

if subtitle_line_mode != "1":

    print("2行字幕設定のためスキップしました。")

else:

    print("=== 1行字幕へ分割中 ===")

    new_segments = []

    new_id = 1


    for seg in formatted_segments:

        text = seg["text"]

        chars = seg.get("chars", [])


        # 改行が無ければそのまま
        if "\n" not in text:

            seg["id"] = new_id

            new_segments.append(seg)

            new_id += 1

            continue


        line1, line2 = text.split("\n", 1)


        split_pos = len(line1)


        if chars and split_pos < len(chars):

            split_sec = chars[split_pos]["start"]

            chars1 = chars[:split_pos]

            chars2 = chars[split_pos:]

        else:

            split_sec = (
                seg["start"]
                +
                seg["end"]
            ) / 2

            chars1 = []

            chars2 = []


        new_segments.append(
            {
                "id": new_id,
                "start": seg["start"],
                "end": split_sec,
                "text": line1,
                "chars": chars1
            }
        )

        new_id += 1


        new_segments.append(
            {
                "id": new_id,
                "start": split_sec,
                "end": seg["end"],
                "text": line2,
                "chars": chars2
            }
        )

        new_id += 1


    formatted_segments = new_segments

    current_srt = chunks_to_srt_text(
        formatted_segments
    )

    print("=== 1行字幕へ変換完了 ===")

    print(current_srt)
```

## ⑮ 誤字・表記統一・自然な校正（Gemini）

Geminiを使用して、文字起こし結果の誤字修正・表記統一・自然な日本語への校正を行います。
長尺素材では処理時間が長くなる場合があります。
一時的なエラーが発生した場合は自動で再試行しますが、**API利用上限に達した場合は時間を置いてから再実行してください。**

```python
if use_gemini == "1":

    # 誤字・表記統一・自然な校正（Gemini）
    # 本文分離方式

    import concurrent.futures
    import time
    import re


    def build_gemini_prompt(text):

        return f"""
    以下の字幕文章を校正してください。

    【目的】
    ・誤字脱字の修正
    ・漢字、かな表記の統一
    ・自然な日本語への軽微な修正

    【禁止事項】
    ・意味を変更しない
    ・文章を要約しない
    ・文章量を大幅に減らさない
    ・フィラー削除をしない

    【重要】
    番号は絶対に変更しないでください。
    文章のみ修正してください。

    出力は番号付き文章のみ。

    【入力】

    {text}
    """


    def extract_text_for_gemini(chunks):

        """
        Geminiへ送る本文だけを作成
        タイムコードは送らない
        """

        lines = []

        for i, c in enumerate(chunks, start=1):

            lines.append(
                f"{i}\n{c['text']}"
            )

        return "\n\n".join(lines)



    def parse_gemini_result(text):

        """
        Gemini結果から番号ごとの本文を取得
        """

        blocks = re.split(
            r"\n\s*\n",
            text.strip()
        )

        result = {}

        for block in blocks:

            lines = block.strip().split("\n")

            if len(lines) >= 2:

                try:
                    num = int(lines[0].strip())

                    body = "".join(
                        lines[1:]
                    )

                    result[num] = body

                except:

                    pass

        return result



    def generate_with_retry(
            prompt,
            max_retries=5,
            timeout_sec=120):


        wait = 10


        for attempt in range(1, max_retries + 1):

            try:

                with concurrent.futures.ThreadPoolExecutor(max_workers=1) as executor:

                    future = executor.submit(
                        model.generate_content,
                        prompt
                    )

                    response = future.result(
                        timeout=timeout_sec
                    )


                if not response.text:

                    raise Exception(
                        "empty response"
                    )


                return response.text



            except Exception as e:


                error = str(e)

                print(
                    f"Geminiエラー {attempt}/{max_retries}"
                )

                print(error)



                # 429は即停止

                if (
                    "429" in error
                    or "TooManyRequests" in error
                    or "quota" in error.lower()
                ):

                    print("")
                    print("==============================")
                    print("⚠️ Gemini API無料枠上限です")
                    print("処理を停止します")
                    print("==============================")

                    raise RuntimeError(
                        "Gemini quota exceeded"
                    )



                # 503 / Timeoutのみリトライ

                if (
                    "503" in error
                    or isinstance(e, TimeoutError)
                ):

                    if attempt < max_retries:

                        print(
                            f"{wait}秒待機して再試行"
                        )

                        time.sleep(wait)

                        wait = min(
                            wait * 2,
                            300
                        )

                        continue



                raise e




    # =========================
    # メイン処理
    # =========================


    if "processed_segments" not in locals():

        raise NameError(
            "processed_segments がありません。前の処理を先に実行してください。"
        )


    batch_size = 100


    batches = [
        processed_segments[i:i + batch_size]

        for i in range(
            0,
            len(processed_segments),
            batch_size
        )
    ]


    print(
        f"{len(batches)}個のバッチに分割しました"
    )



    new_segments = []


    for idx, batch in enumerate(
            batches,
            start=1):


        print(
            f"バッチ {idx}/{len(batches)} 処理中..."
        )


        # 本文だけ抽出

        gemini_text = extract_text_for_gemini(
            batch
        )


        prompt = build_gemini_prompt(
            gemini_text
        )


        result = generate_with_retry(
            prompt
        )


        # 番号別に戻す

        corrected = parse_gemini_result(
            result
        )



        for i, c in enumerate(batch, start=1):

            new_chunk = c.copy()


            if i in corrected:

                new_chunk["text"] = corrected[i]


            else:

                print(
                    f"⚠️ {i}番の戻し失敗。元文章維持"
                )


            new_segments.append(
                new_chunk
            )



        # 連続送信防止

        if idx < len(batches):

            time.sleep(3)



    processed_segments = new_segments


    current_srt = chunks_to_srt_text(
        processed_segments
    )


    print("")
    print("=== Gemini校正完了 ===")
    print(current_srt)

else:

    print("Gemini校正をスキップしました。")

    current_srt = chunks_to_srt_text(
        processed_segments
    )
```

## ⑯ タイムコード復元・SRT検証

```python
# タイムコード復元・SRT検証
# Gemini校正後の本文を元のSRT形式に戻し、安全確認を行うセル

import re

print("=== タイムコード復元・SRT検証を開始します ===\n")


# ⑨の結果確認
if 'current_srt' not in locals():
    raise NameError(
        "Gemini校正後の SRTデータ (current_srt) が見つかりません。前のセルの実行を確認してください。"
    )


# ==========================================
# SRT検証用関数
# ==========================================


def validate_timecode(time_line):
    """
    タイムコード形式チェック
    """

    pattern = (
        r"^\d{2}:\d{2}:\d{2},\d{3}"
        r"\s-->\s"
        r"\d{2}:\d{2}:\d{2},\d{3}$"
    )

    return re.match(pattern, time_line) is not None



def rebuild_srt_numbering(srt_text):
    """
    ブロック番号を振り直し、
    空ブロックを除去する
    """

    blocks = srt_text.strip().split("\n\n")

    new_blocks = []
    new_id = 1

    error_count = 0


    for block in blocks:

        lines = [
            line.strip()
            for line in block.split("\n")
            if line.strip()
        ]


        # 番号・タイムコード・本文がないブロックは削除
        if len(lines) < 3:
            continue


        time_line = lines[1]


        # タイムコード異常
        if not validate_timecode(time_line):
            print(
                f"⚠️ タイムコード形式異常: {time_line}"
            )
            error_count += 1
            continue


        text = "\n".join(lines[2:])


        # 本文空白チェック
        if not text.strip():
            continue


        new_blocks.append(
            f"{new_id}\n"
            f"{time_line}\n"
            f"{text}"
        )

        new_id += 1


    return (
        "\n\n".join(new_blocks),
        new_id - 1,
        error_count
    )


# ==========================================
# 実行
# ==========================================


# ==========================================
# processed_segmentsを維持したまま検証
# ==========================================

if "processed_segments" not in locals():

    raise NameError(
        "processed_segments がありません。"
    )


validated_segments = []


error_count = 0


for i, seg in enumerate(
    processed_segments,
    start=1
):

    if not seg.get("text", "").strip():

        continue


    validated_segments.append(
        {
            "id": i,
            "start": seg["start"],
            "end": seg["end"],
            "text": seg["text"],
            "chars": seg.get(
                "chars",
                []
            )
        }
    )


processed_segments = validated_segments


block_count = len(processed_segments)


current_srt = chunks_to_srt_text(
    processed_segments
)


print("=== SRT検証完了 ===")
print(f"総ブロック数: {block_count} 件")


if error_count > 0:
    print(
        f"⚠️ タイムコード異常 {error_count} 件を検出しました"
    )
else:
    print(
        "✅ タイムコード正常"
    )


print(
    f"出力変数名: current_srt (文字数: {len(current_srt)})"
)


print("\n--- SRTプレビュー ---")
print(current_srt)
```

## ⑰ 最低表示時間の確保

```python
# ==========================================
# 最低表示時間の確保
#
# 最終分割後の表示時間補正
# chars保持
# ==========================================


print("=== 字幕表示時間調整を開始します ===\n")



if "formatted_segments" not in locals():

    raise NameError(
        "formatted_segments が見つかりません。"
        "前の処理セルを先に実行してください。"
    )



# ==========================================
# 最低表示時間調整
# ==========================================

def enforce_min_duration_by_chars(
    segments,
    sec_per_char=0.3,
    min_floor=0.8,
    gap=0.0
):

    """
    最終字幕の表示時間補正

    ・文字数から必要表示時間を計算
    ・不足時のみendを延長
    ・charsは保持
    ・次字幕との重複は禁止
    """



    adjusted = []



    for i, block in enumerate(
        segments
    ):


        new_block = block.copy()



        text = (
            block["text"]
            .replace("\n", "")
        )



        char_count = len(
            text
        )



        duration = (
            block["end"]
            -
            block["start"]
        )



        required = max(
            char_count * sec_per_char,
            min_floor
        )



        if duration < required:


            desired_end = (
                block["start"]
                +
                required
            )



            # --------------------------
            # 次字幕との重複防止
            # --------------------------

            if i + 1 < len(segments):


                next_start = (
                    segments[i + 1]["start"]
                )


                new_block["end"] = min(
                    desired_end,
                    next_start - gap
                )


            else:


                new_block["end"] = (
                    desired_end
                )



            # --------------------------
            # 異常防止
            # --------------------------

            if (
                new_block["end"]
                <=
                new_block["start"]
            ):

                new_block["end"] = (
                    new_block["start"]
                    +
                    0.1
                )



        adjusted.append(
            new_block
        )


    return adjusted



# ==========================================
# 実行
# ==========================================


formatted_segments = enforce_min_duration_by_chars(
    formatted_segments
)



# ==========================================
# SRT生成
# ==========================================

current_srt = chunks_to_srt_text(
    formatted_segments
)



print(
    "=== 表示時間調整完了 ==="
)


print(
    f"ブロック数: {len(formatted_segments)}"
)


print(
    f"文字数: {len(current_srt)}"
)


print(
    "\n--- 調整後プレビュー ---"
)


print(
    current_srt
)
```
